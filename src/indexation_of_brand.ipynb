```python
import os
import pandas as pd

import data_preparation as mdp
from mercari_config import MercariConfig
```

```python
train_data = mdp.load_data(MercariConfig.TRAINING_SET_PREP_FILE)
val_data = mdp.load_data(MercariConfig.VALIDATION_SET_PREP_FILE)
```

```python
brands = train_data[['brand_name', 'name']]
brands = brands.groupby("brand_name").count()

brands.columns = ['count']
```

```python
brands['brand_id'] = [i for i in range(MercariConfig.WORD_I, len(brands) + MercariConfig.WORD_I)]

brands.at[MercariConfig.PAD, 'brand_id'] = MercariConfig.PAD_I
brands.at[MercariConfig.START, 'brand_id'] = MercariConfig.START_I
brands.at[MercariConfig.OOV, 'brand_id'] = MercariConfig.OOV_I
brands.at[MercariConfig.REMOVED_PRICE, 'brand_id'] = MercariConfig.REMOVED_PRICE_I
brands.at[MercariConfig.EMPTY_NAME, 'brand_id'] = MercariConfig.EMPTY_NAME_I
brands.at[MercariConfig.EMPTY_CAT, 'brand_id'] = MercariConfig.EMPTY_CAT_I
brands.at[MercariConfig.EMPTY_BRAND, 'brand_id'] = MercariConfig.EMPTY_BRAND_I
brands.at[MercariConfig.EMPTY_DESC, 'brand_id'] = MercariConfig.EMPTY_DESC_I

brands['count'].fillna(value=0, inplace=True)
brands.sort_values(by='brand_id', inplace=True)
brands = brands.astype('int32')
```

```python
def index_brand(data):
    data['brand_id'] = 0

    data_len = len(data)

    progress = 0

    row_iterator = data.iterrows()

    for index, row in row_iterator:
        brand_nm = row['brand_name']

        if brand_nm in brands.index:
            data.at[index, 'brand_id'] = brands.at[brand_nm, 'brand_id']
        else:
            data.at[index, 'brand_id'] = MercariConfig.EMPTY_BRAND_I


        progress += 1

        if not progress % 10000:
            print("Progress: %3.2f" % (progress * 100.0 / data_len))
```

```python
%%time

index_brand(train_data)
```

```python
%%time

index_brand(val_data)
```
