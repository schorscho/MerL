```python
import os
import pandas as pd

import data_preparation as mdp
from mercari_config import MercariConfig
```

```python
df = mdp.load_data(MercariConfig.TRAINING_SET_PREP_FILE)
```

```python
brands = df.groupby("brand_name").count()
```

```python
brands['brand_id'] = [i for i in range(len(brands))]
```

```python
%%time

df['brand_id'] = 0

data_len = len(df)

progress = 0

row_iterator = df.iterrows()

for index, row in row_iterator:
    df.at[index, 'brand_id'] = brands.at[row['brand_name'], 'brand_id']
    
    progress += 1

    if not progress % 10000:
        print("Progress: %3.2f" % (progress * 100.0 / data_len))
```
